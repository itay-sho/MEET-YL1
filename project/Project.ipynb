```python
import csv
```

importing the needed libraries

```python
data = open('environmental.csv','rb')
reader = csv.reader(data);
index = 0
ozone = []
radiation = []
temperature = []
wind = []
for row in reader:
    # if index is 0 it means that this is the row names
    if index != 0:
        # changing the strings into floats
        row = map(float, row)
        # adding the params to the right list
        ozone.append(row[0])
        radiation.append(row[1])
        temperature.append(row[2])
        wind.append(row[3])
    # counting up the index
    index += 1
# closing the file
data.close()
```
